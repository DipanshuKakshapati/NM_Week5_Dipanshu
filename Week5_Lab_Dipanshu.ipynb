### Multiple Application Trapezoidal Rule

User can input their preferred function value f(x), lower limit (a), upper limit (b), and number of segments (n) between the intervals of a and b.

```python
import numpy as np
import matplotlib.pyplot as plt

def trapezoidal_rule_multiple(f, a, b, n):
    h = (b - a) / n
    x = np.linspace(a, b, n+1)
    fx = f(x)
    area = h * (0.5 * fx[0] + 0.5 * fx[-1] + sum(fx[1:-1]))
    plot_function(f, a, b, x, fx, h, area)
    return area

def plot_function(f, a, b, x, fx, h, area):
    # Plot the function
    x_fine = np.linspace(a, b, 1000)  # a finer grid for a smooth curve of the function
    y_fine = f(x_fine)
    plt.plot(x_fine, y_fine, 'b-', label='Function f(x)')

    # Plot the trapezoids
    for i in range(len(x)-1):
        plt.fill_between([x[i], x[i+1]], [fx[i], fx[i+1]], color='lightgray', alpha=0.5)
        plt.plot([x[i], x[i+1]], [fx[i], fx[i+1]], 'r-', label='Trapezoid' if i == 0 else "")

    # Highlight the approximated value on the plot
    plt.axhline(y=area, color='green', linestyle='--', label=f'Approximated value = {area}')

    # Formatting the plot
    plt.title('Trapezoidal Rule Approximation')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    print("Multiple-Segment Trapezoidal Rule Calculator")
    
    # Get user input for the function
    func_str = input("Enter the function to integrate (e.g., 'x**2'): ")
    f = lambda x: eval(func_str)
    
    # Get user input for the limits of integration
    a = float(input("Enter the lower limit of integration (a): "))
    b = float(input("Enter the upper limit of integration (b): "))
    
    # Get user input for the number of segments
    n = int(input("Enter the number of segments (n): "))
    
    # Calculate the integral and plot the function
    result = trapezoidal_rule_multiple(f, a, b, n)
    print(f"The approximate value of the integral is: {result}")

if __name__ == "__main__":
    main()
```

### Mulitple Application of Simpson's 1/3 Rule

User can input their preferred function value f(x), lower limit (a), upper limit (b), and number of segments (n) between the intervals of a and b.

```python
import numpy as np
import matplotlib.pyplot as plt

def simpsons_rule_multiple_13(f, a, b, n):
    if n % 2 == 1:  # Ensure n is even
        n += 1
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    fx = f(x)
    
    # Simpson's 1/3 rule calculation
    integral = (h / 3) * (fx[0] + 2 * sum(fx[2:-1:2]) + 4 * sum(fx[1:n:2]) + fx[-1])
    return integral, x, fx

def plot_function(f, a, b, x, fx, integral):
    x_fine = np.linspace(a, b, 1000)
    y_fine = f(x_fine)
    plt.plot(x_fine, y_fine, 'b-', label='Function f(x)')
    
    # Plot the approximation segments
    for i in range(1, len(x)-1, 2):
        plt.fill_between(x[i-1:i+2], fx[i-1:i+2], color='lightgrey', alpha=0.5)
        plt.plot(x[i-1:i+2], fx[i-1:i+2], 'r-', label='Simpson\'s 1/3 segments' if i == 1 else "")
        
     # Highlight the approximated value on the plot
    plt.axhline(y=integral, color='green', linestyle='--', label=f'Approximated value = {integral}')
    
    plt.title('Function and Simpson\'s 1/3 Rule Approximation')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.show()

def main():
    print("Multiple-Application Simpson's Rule Calculator")
    
    func_str = input("Enter the function to integrate (e.g., 'x**2'): ")
    f = lambda x: eval(func_str)
    
    a = float(input("Enter the lower limit of integration (a): "))
    b = float(input("Enter the upper limit of integration (b): "))
    n = int(input("Enter the number of segments (n): "))
    
    integral, x, fx = simpsons_rule_multiple_13(f, a, b, n)
    print(f"The approximate value of the integral is: {integral}")
    
    plot_function(f, a, b, x, fx, integral)

if __name__ == "__main__":
    main()
```

### Mulitple Application of Simpson's 3/8 Rule

User can input their preferred function value f(x), lower limit (a), upper limit (b), and number of segments (n) between the intervals of a and b.

```python
import numpy as np
import matplotlib.pyplot as plt

def simpsons_38_rule_multiple(f, a, b, n):
    # Ensure n is a multiple of three
    if n % 3 != 0:
        n += 3 - (n % 3)
    
    h = (b - a) / n
    x = np.linspace(a, b, n + 1)
    fx = f(x)
    
    # Simpson's 3/8 rule calculation
    integral = 3 * h / 8 * (fx[0] + fx[-1] + 3 * sum(fx[1:-1:3]) + 3 * sum(fx[2:-1:3]) + 2 * sum(fx[3:-3:3]))
    return integral, x, fx

def plot_function(f, a, b, x, fx, integral):  # Modified to accept 'integral' as an argument
    x_fine = np.linspace(a, b, 1000)
    y_fine = f(x_fine)
    plt.plot(x_fine, y_fine, 'b-', label='Function f(x)')

    # Highlight the segments used for Simpson's 3/8 rule
    for i in range(0, len(x) - 1, 3):
        if i + 3 < len(x):
            plt.fill_between(x[i:i+4], fx[i:i+4], color='lightgrey', alpha=0.5)
            plt.plot(x[i:i+4], fx[i:i+4], 'r-', label='Simpson\'s 3/8 segments' if i == 0 else "")
    
    # Highlight the approximated value on the plot
    plt.axhline(y=integral, color='green', linestyle='--', label=f'Approximated value = {integral:.4f}')
    
    plt.title('Function and Simpson\'s 3/8 Rule Approximation')
    plt.xlabel('x')
    plt.ylabel('f(x)')
    plt.legend()
    plt.grid(True)
    plt.show()

def main():
    print("Multiple-Application Simpson's 3/8 Rule Calculator")
    
    func_str = input("Enter the function to integrate (e.g., 'x**2'): ")
    f = lambda x: eval(func_str)
    
    a = float(input("Enter the lower limit of integration (a): "))
    b = float(input("Enter the upper limit of integration (b): "))
    n = int(input("Enter the number of segments (n): "))
    
    integral, x, fx = simpsons_38_rule_multiple(f, a, b, n)
    print(f"The approximate value of the integral is: {integral}")
    
    plot_function(f, a, b, x, fx, integral)  # Pass 'integral' to the plotting function

if __name__ == "__main__":
    main()
```
